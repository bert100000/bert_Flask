```python
import requests
```

```python
def get_pm25(sort=False):
    columns,values,datas=None,None,None
    try:
        url = 'https://data.moenv.gov.tw/api/v2/aqx_p_02?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate%20desc&format=JSON'
        resp = requests.get(url)
        datas = resp.json()['records']
        columns = list(datas[0].keys())
        values=[]
        for data in datas:
            data=list(data.values())
            try:
                data[2]=eval(data[2])
                values.append(data)
            except Exception as e:
                print(e)
        if sort:
            values=sorted(values,key=lambda x:x[2],reverse=True)
        
        print(values)
    except Exception as e:
        print(e)

    return columns,values,datas
```

```python
columns, values,datas = get_pm25(sort=True)
```

# 建立資料庫
- 1.創建資料庫
- 2.建立資料表
- 3.寫入不重覆資料
- 4.取最新資料

```python
import sqlite3
```

```python
conn = sqlite3.connect('pm25.db')
cursor = conn.cursor()

conn,cursor
```

```python
datas[0].keys()
```

```python
sqlstr='''
create table if not exists data(
id integer primary key autoincrement,
site text,
county text,
pm25 integer,
datacreationdate datetime,
itemunit text
)
'''
cursor.execute(sqlstr)
conn.commit()
```

### 檢查資料是否存在

### 寫入資料

```python
count = 0
for data in datas:
    data = list(data.values())
    sqlstr = f'select * from data where site="{data[0]}" and datacreationdate="{data[-2]}"'
    result = list(cursor.execute(sqlstr))
    if result ==[]:
        sqlstr=f'insert into data(site,county,pm25,datacreationdate,itemunit) values("{data[0]}","{data[1]}",\
        {data[2]},"{data[3]}","{data[4]}")'
        cursor.execute(sqlstr)
        count+=1
    
conn.commit()
print(f'更新:{count}筆資料')
```
